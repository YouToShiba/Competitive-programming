```python
N, A = map(int,input().split())
x = list(map(int,input().split()))
# dp[[51][102]...[51*N]]
dp = [[0]*((i+1)*51)for i in range(N+1)]
dp[0][0]=1

for n in range(N):
  a=x[n]
  for k in range(n+1,0,-1):
    for s in range(a,k*50+1):
      dp[k][s]=dp[k][s]+dp[k-1][s-a]
print(sum([dp[n][n*A]for n in range(1,N+1)])) 
```
