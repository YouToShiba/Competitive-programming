```python
S = input()
ans = 0

# 全探索
# 1をlen(S)-1ビット左シフト (ex. len(S)=5->10000(10進数で16), -1で1111となる)
# ex) 10010のとき1, 4番目に+を挿入
for i in range(1<<(len(S)-1)):
    S_temp = list(S)
    cnt = 0
    for shift in range(len(S)):
        # &:ビット演算子の方のand
        #print("------")
        #print(i)
        #print(shift)
        if (i & 1<<shift):
            S_temp.insert(shift+cnt+1,'+')
            cnt += 1
    # print(S_temp)
    # eval:文字列で表された式を計算
    ans += eval(''.join(S_temp))
print(ans)
```
